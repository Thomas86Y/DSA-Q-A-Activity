```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.neural_network import MLPClassifier
```

```python
mush = pd.read_csv("mushrooms.csv")
```

```python
mush
```

```python
mush.isnull().sum()
```

```python
# Visualization
```

```python
sns.countplot(x="class",data = mush)
plt.title("Ratio of Poisonous and Edible mushrooms")
```

```python
# Preprocessing
```

```python
mappings = list()
en = LabelEncoder()
for column in range(len(mush.columns)):
    mush[mush.columns[column]] = en.fit_transform(mush[mush.columns[column]])
    map_dict = {index: label for index,label in enumerate(en.classes_)}
    mappings.append(map_dict)
```

```python
mush
```

```python
mappings
```

```python
y = mush["class"]
X = mush.drop("class", axis =1)
```

```python
# Standard Scaling
```

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8,random_state=42)
```

```python
# Model Selection
```

```python
log_model = LogisticRegression()
svm_model = SVC(C=1.0,kernel = "rbf")
nn_model = MLPClassifier(hidden_layer_sizes=(128,128))
```

```python
np.sum(y) / len(y)
```

48% positives are there

```python
log_model.fit(X_train,y_train)
svm_model.fit(X_train,y_train)
nn_model.fit(X_train,y_train)
```

```python
print(f"---Logestic Regression: {log_model.score(X_test,y_test)}")
print(f"Support Vector Machine: {svm_model.score(X_test,y_test)}")
print(f"---------Nural Network: {nn_model.score(X_test,y_test)}")
```

```python
# Applying GaussianNB Classifier
```

```python
classifier = GaussianNB()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
# Applying BernoulliNB Classifier
```

```python
classifier = BernoulliNB()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```
