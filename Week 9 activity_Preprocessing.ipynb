# 7. Check and handle missing values in the dataset.


```python
emp[["age","income","gender"]]
```

age with mean,  income with median ,  gender with mode  (filling the null values)

```python
emp['age']    = emp["age"].fillna(emp["age"].mean())
emp['income'] = emp["income"].fillna(emp["income"].median())
emp['gender'] = emp["gender"].fillna(emp["gender"].mode().loc[0])
```

# 8. Check the outliers and handle outliers in performance score using Percentiles.

```python
import matplotlib.pyplot as plt
plt.boxplot(emp["performance_score"])
plt.title("Checking outliers")
```

```python
Q1 = np.percentile(emp["performance_score"], 25)
Q2 = np.percentile(emp["performance_score"], 50)
Q3 = np.percentile(emp["performance_score"], 75)
IQR = Q3 - Q1
low_lim = Q1 - 1.5*IQR
Up_lim  = Q3 + 1.5*IQR
```

```python
print(low_lim)
print(Up_lim)
```

```python
Outlier= emp[(emp['performance_score'] < low_lim ) | (emp['performance_score'] > Up_lim)]
Outlier
```

```python
index1 = emp["performance_score"] < low_lim
emp.loc[index1,'performance_score'] = low_lim
```

```python
emp
```

```python
plt.boxplot(emp["performance_score"])
plt.title("Checking outliers")
```
