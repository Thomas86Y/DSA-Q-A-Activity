```python
import pandas as pd
import seaborn as sns
```

```python
sp = pd.read_csv("StudentsPerformance.csv")
```

```python
sp
```

# 1. Find out how many males and females participated in the test.

```python
sns.countplot(x= "gender",data = sp)
print("Number of Male participated in the test :" ,sp["gender"].value_counts()['male'])
print("Number of Female participated in the test :" ,sp["gender"].value_counts()['female'])      
```

# 2. What do you think about the students' parental level of education?


```python
import matplotlib.pyplot as plt
sns.countplot(x = "parental level of education",data = sp)
plt.title("parental level of education")
plt.xticks(rotation = 90)
sp['parental level of education'].value_counts()
```

# 3. Who scores the most on average for math, reading and writing based on
● Gender


```python
sp1=sp.groupby('gender')['gender','math score','reading score','writing score'].mean()
print('Average wrt gender is :',sp1)
fig=plt.subplots(figsize=(12,5))
plt.title("Age score wrt gender")
plt.xticks(rotation = 90)
ax=sns.lineplot(data=sp1,markers=True)
ax.set(ylabel="Average score")
```

● Test preparation course

```python
sp2=sp.groupby('test preparation course')['test preparation course','math score','reading score','writing score'].mean()
print('Average wrt test preparation course is :',sp2)
fig=plt.subplots(figsize=(12,5))
plt.title("Age score wrt test preparation course")
plt.xticks(rotation = 90)
ax=sns.lineplot(data=sp2,markers=True)
ax.set(ylabel="Average score")

```

# 4. What do you think about the scoring variation for math, reading and writing
based on
● Gender

```python
s_v=sp.groupby('gender')['math score','reading score','writing score'].std()/sp.groupby('gender')['math score','reading score','writing score'].mean()
print("Score variation wrt Gender :\n",s_v)
fig=plt.subplots(figsize=(12,5))
plt.title("Age score wrt gender")
plt.xticks(rotation = 90)
ax=sns.lineplot(data=s_v,markers=True)
ax.set(ylabel="Average score")
```

● Test preparation course

```python
s_v1=sp.groupby('test preparation course')['math score','reading score','writing score'].std()/sp.groupby('test preparation course')['math score','reading score','writing score'].mean()
print("Score variation wrt test preparation course :\n",s_v1)
fig=plt.subplots(figsize=(12,5))
plt.title("Age score wrt test preparation courser")
plt.xticks(rotation = 90)
ax=sns.lineplot(data=s_v1,markers=True)
ax.set(ylabel="Average score")
```
