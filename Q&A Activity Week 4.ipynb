# 1. Compare churn count with respect to gender.


```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
sns.countplot(x = "gender",hue = "Churn",data = tc)
plt.title("Churn Summary wrt gender",fontsize=17)
```

# 2. Find out how many female senior citizens there in the dataset

```python
sns.countplot(x = 'gender',hue = 'SeniorCitizen',data = tc)
plt.title('Female Senior Citizens',fontsize=17)



```

# 3. Compare 'tenure' with 'Total Charges'

```python
sns.scatterplot(data=tc, x="tenure",y="TotalCharges").set(title='Total charges vs Tenure scatter plot')
```

# 4. Find out which contract is preferred by the senior citizen.

```python
scz=tc.copy()
scz.loc[scz['SeniorCitizen']==0] 
sns.catplot(x='Contract', kind="count", data=scz).set(title=' preference contract for Senior citizens')
```
